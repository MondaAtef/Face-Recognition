## data_train

```python
import os
import cv2
import numpy as np
import face_recognition

# Directory path containing the image files
directory = r"C:\Users\Office\Downloads\pattern_project\Original Images\train"

# Lists to store the encoded faces and corresponding classes
train_encodes = []
train_classes = []

# Loop through each subfolder (person) in the directory
for person_folder in os.listdir(directory):
    person_folder_path = os.path.join(directory, person_folder)
    if os.path.isdir(person_folder_path):
        # Iterate over each image file in the person's folder
        for image_name in os.listdir(person_folder_path):
            # Construct the full file path
            image_path = os.path.join(person_folder_path, image_name)
            
            # Read the image using OpenCV
            img = cv2.imread(image_path)
            if img is not None:
                # Convert the image to RGB color space
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                # Extract face encodings from the image
                face_encodings = face_recognition.face_encodings(img)
                if len(face_encodings) > 0:
                    # If a face is detected, append its encoding and corresponding label to the lists
                    train_encodes.append(face_encodings[0])  # Assuming only one face per image
                    train_classes.append(person_folder)

print(train_classes)
print(train_encodes)
```

## compare a list of known face encodings with a single unknown face


```python
import os
import cv2
import face_recognition
import matplotlib.pyplot as plt

# Set the number of rows and columns for subplots
num_rows = 3
num_cols = 4

plt.figure(1, figsize=(10, 5))

# Path to the test images
test_path = r"C:\Users\Office\Downloads\pattern_project\test"
files_in_test_path = os.listdir(test_path)

k = 0
num_files_in_test_path = len(files_in_test_path)
for i, image_name in enumerate(os.listdir(test_path)):
    img = cv2.imread(os.path.join(test_path, image_name))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Get face encodings
    face_encodings = face_recognition.face_encodings(img)
    
    if len(face_encodings) > 0:
        img_encoding = face_encodings[0]  # Take the first face if multiple faces are detected
        result = face_recognition.compare_faces(train_encodes, img_encoding)
       
        face = "Unknown"
        for j, Class in enumerate(result):
            if Class:
                face = train_classes[j]
                break
    else:
        face = "No Face Detected"
        k+=1

    # Calculate the subplot position based on the row and column
   
    
    plt.subplot(num_rows, num_cols, i+1)

    plt.axis("off")
    plt.title(face)
    plt.imshow(img)
plt.tight_layout()
plt.show()

acc = ((num_files_in_test_path - k )/ num_files_in_test_path)*100

print("accuracy = ",acc,"%")
```

## face recogntion in video


```python
# Open the video file
cap = cv2.VideoCapture('C:/Users/Office/Downloads/pattern_project/barak.mp4')

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    frame = cv2.resize(frame, (1080, 500)) 
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Get face locations and encodings from the frame
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

   
    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
     
        result = face_recognition.compare_faces(train_encodes, face_encoding)

       
        face = "Unknown"

        # Find the corresponding label for the recognized face
        for j, is_match in enumerate(result):
            if is_match:
                face = train_classes[j]
                break

        
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
        cv2.putText(frame, face, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

    
    cv2.imshow("test window", frame)
    cv2.imwrite("C:/Users/Office/Downloads/pattern_project/test/numberplate.jpg", frame)
    if cv2.waitKey(5) & 0xFF == 27:
        break

cap.release()   
cv2.destroyAllWindows()
```
